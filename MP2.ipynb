```python
def split_matrix2(demo):
    rows=0
    cols=0
    arr = np.zeros((6,6), dtype=str)
    for i in demo:
        arr[rows][cols]=i
        cols+=1
        if cols==6:
            rows+=1
            cols=0
            if rows==6:break
    return arr

```

```python
def check_neighboors_down(state_space, row, column):
    result = ['null',-1]

    # check for lower limit
    if row>3: 
        return result
    
    else:
        #going through every cell down and trying to find a car
        for i in range(row+1, 5):
            #if below cell is empty -> continue
            if state_space[i][column]=='.': continue
            #if below cell is lettr -> test for a car
            else: 
                car_letter=state_space[i][column]
                car_length=weights[car_letter]
                #test for a car(next letter the same or not)
                if (state_space[i+1][column]==car_letter): #car found
                    result[0]=car_letter
                    result[1]=i
                    break
                #no car found
                else: break
    return result
```

```python
def check_neighboors_up(state_space, row, column):
    result = ['null',-1]

    # check for upper limit
    if row<2: 
        return result
    
    else:
        #going through every cell up and trying to find a car
        for i in reversed(range(0,row)):
            #if above cell is empty -> continue
            if state_space[i][column]=='.': continue
            #if above cell is lettr -> test for a car
            else: 
                car_letter=state_space[i][column]
                car_length=weights[car_letter]
                #test for a car(next letter the same or not)
                if (state_space[i-1][column]==car_letter): #car found
                    result[0]=car_letter
                    result[1]=i
                    break
                #no car found
                else: break
    return result
```

```python
def check_neighboors_left(state_space, row, column):
    result = ['null',-1]

    # check for left limit
    if column<2: 
        return result
    
    else:
        #going through every cell left and trying to find a car
        for i in reversed(range(0,column)):
            #if left cell is empty -> continue
            if state_space[row][i]=='.': continue
            #if left cell is lettr -> test for a car
            else: 
                car_letter=state_space[row][i]
                car_length=weights[car_letter]
                #test for a car(next letter the same or not)
                if (state_space[row][i-1]==car_letter): #car found
                    result[0]=car_letter
                    result[1]=i
                    break
                #no car found
                else: break
    return result
```

```python
def check_neighboors_right(state_space, row, column):
    result = ['null',-1]

    # check for right limit
    if column>3: 
        return result
    
    else:
        #going through every cell right and trying to find a car
        for i in range(column+1,5):
            #if right cell is empty -> continue
            if state_space[row][i]=='.': continue
            #if right cell is lettr -> test for a car
            else: 
                car_letter=state_space[row][i]
                car_length=weights[car_letter]
                #test for a car(next letter the same or not)
                if (state_space[row][i+1]==car_letter): #car found
                    result[0]=car_letter
                    result[1]=i
                    break
                #no car found
                else: break
    return result
```

```python
def check_fuel(ls):
    fuelList = {}
    cars = []
    for char in ls.split()[0]:
        if char != '.' and char not in cars:
            cars.append(char)
    customFuel = ls.split()[1:]
    if customFuel != []:
        for j in range(len(customFuel)):
            letter = customFuel[j][0]
            fuel = int(customFuel[j][1])
            fuelList[letter] = fuel
    for car in cars:
        if car not in fuelList.keys():
            fuelList[car] = 100
    return fuelList
    
```

```python
def H1(s):   #creation of heuristic "The number of blocking vehicles" to pass each node to
    #this will take the passed child state of 'A' and summ the number of blocking vehicles and return that value as the heuristic
    transpose2 = np.argwhere(s == "A")
    transpose3=np.argwhere(s==s)
    #goal=transpose3[17]
    counter=0
    next_col=(transpose2[1][1]+1)
    #transpose2[1][1] gives the final column location of the second A in AA and transpose[1][0] gives the row location
    check = False
    
    
    #check if slots between AA and goal state contain cars or not (".")
    
    #this will loop through transpose3 in the specific row and column +1 to the right of AA

    for i in (transpose2[1][0],transpose2[1][0]):   #row of AA
        if  check == False:
            if (transpose3[1][1])+1<=5: #this ensures we do not go out of bounds for our check
                for j in range(next_col,6): #one column after AA to start checking
                    #horizontal check
                    if j+1<5:
                        if (s[i][j]!="." and s[i][j+1]==s[i][j]):   #a horizontal car has been found taking up more than one space
                            counter+=0
                            
                            #vertical check
                        if (s[i][j]!=".") and s[i][j+1]!=s[i][j]:  #a vertical car has been found taking up one spot from the horizontal goal
                            counter+=1
                    else:
                        if j+1<=5:
                            if (s[i][j]!=".") and s[i][j+1]!=s[i][j]:  #a vertical car has been found taking up one spot from the horizontal goal
                                    counter+=1

                    if j==5:
                        if (s[i][j]!="." and s[i][j-1]!=s[i][j]):  #a vertical car has been found taking up one spot from the horizontal goal
                                counter+=1
                                check = True
                                break
                    if j==5:
                        if (s[i][j]=="."): 
                            check = True
                            break
                        
            
    #print("Number of Cars blocking AA from exit is: ",counter)
    return  counter
```
